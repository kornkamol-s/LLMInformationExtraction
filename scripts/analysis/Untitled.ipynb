```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
toc = pd.read_csv('C:/Users/KornkamolSampaongern/LLMInformationExtraction/data/intermediate/analysis/table_of_contents.csv')
```

```python
G = nx.Graph()

for section, group in toc.groupby('section'):
    ids = group['id'].tolist()
    for i in range(len(ids)):
        for j in range(i + 1, len(ids)):
            G.add_edge(ids[i], ids[j], section=section)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', edge_color='gray')

plt.title('Graph of Linked IDs by Section')
plt.show()
```
